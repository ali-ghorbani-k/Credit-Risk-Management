```python
bureau = pd.read_csv('bureau.csv')
bureau.head()
```

```python
bureau.shape
```

```python
bureau = bureau.drop_duplicates()
bureau.shape
```

```python
bureau.select_dtypes(include = object).isnull().sum()
```

```python
bureau.select_dtypes(exclude = object).isnull().sum()
```

```python
bureau.fillna(0, inplace =True)
bureau.isnull().sum()
```

```python
bureau['Duration_planned'] = bureau['DAYS_CREDIT_ENDDATE'] - bureau['DAYS_CREDIT'] 
bureau['Duration_real']    = bureau['DAYS_ENDDATE_FACT'] - bureau['DAYS_CREDIT']
```

```python
# Nagative days_early_paid means: prolonged bureau
bureau['Days_early_paidoff'] =  bureau['DAYS_CREDIT_ENDDATE'] - bureau['DAYS_ENDDATE_FACT']
```

```python
# Weighted sum of credit a person borrowed (per days) (later aggregation: Weighted sume of credit debt per day per each credit)
bureau['AMT_weightdebt_duration'] = bureau['AMT_CREDIT_SUM_DEBT']/bureau['Duration_planned']
```

```python
# 'AMT_CREDIT_SUM_OVERDUE': Current amount overdue on credit 
bureau['AMT_Overdue_duration'] = bureau['AMT_CREDIT_SUM_OVERDUE']/bureau['Duration_planned'] 
```

```python
# Maximal amount overdue so far 
bureau['AMT_Maxoverdue_duration'] = bureau['AMT_CREDIT_MAX_OVERDUE']/bureau['Duration_planned'] 
```

```python
# Default credit day overdue more than 270 days
bureau['Defaulted'] = np.where(bureau['CREDIT_DAY_OVERDUE'] > 270, 1, 0)
bureau['AMT_defaulted'] = bureau['Defaulted']*bureau['AMT_CREDIT_SUM_DEBT']
# np.count_nonzero(bureau['Default'])
```

```python
# number of current credit active: sum of credit active atttibute
mapping = {'Closed':0, 'Active': 1}
bureau['CREDIT_ACTIVE'] = bureau['CREDIT_ACTIVE'].map(mapping)
bureau['CREDIT_ACTIVE']
```

```python
stats = bureau.sort_values('DAYS_CREDIT', ascending = False).groupby('SK_ID_CURR') \
        .agg({'AMT_CREDIT_SUM_DEBT': ['count' ,'sum', 'mean', 'first'], 
              'AMT_weightdebt_duration': ['sum', 'mean', 'first'], 
              'AMT_Overdue_duration': [ 'sum', 'mean', 'first'], 
              'AMT_Maxoverdue_duration': ['mean', 'first'], 
              'Days_early_paidoff': [ 'sum', 'mean', 'first'],
              'Defaulted':[ 'sum', 'mean', 'first'], 
              'AMT_defaulted':[ 'sum', 'mean', 'first'],
              'DAYS_CREDIT': 'first', 'CREDIT_ACTIVE': 'sum'}) 
stats
```

```python
bureau.groupby('CREDIT_TYPE')['AMT_CREDIT_SUM_DEBT'].mean().sort_values(ascending = False)
```

```python
# Filter 6 categories of loans
Filter = (bureau['CREDIT_TYPE'] == 'Mortgage')  | (bureau['CREDIT_TYPE'] == 'Real state loan') \
         | (bureau['CREDIT_TYPE'] == 'Car loan') \
         | (bureau['CREDIT_TYPE'] == 'Loan for business development') \
         | (bureau['CREDIT_TYPE'] == 'Loan for purchase of equipment') \
         | (bureau['CREDIT_TYPE'] == 'Cash loan (non-earmarked)')\
         | (bureau['CREDIT_TYPE'] == 'Loan for working capital replenishment') 
btype = bureau[Filter].copy()
btype
```

```python
btype['Defaulted'] = np.where(btype['CREDIT_DAY_OVERDUE'] > 270, 1, 0)
btype['AMT_defaulted'] = btype['Defaulted']*btype['AMT_CREDIT_SUM_DEBT']
```

```python
# first(DAYS_CREDIT) is the last credit open for each bureau credit, last(Days_Credit) is the first day 
# first/last credit bureau open, close,times credit deafult, amount credit defaulted 
bt_stat  = btype.sort_values('DAYS_CREDIT', ascending = False).groupby(['SK_ID_CURR','CREDIT_TYPE'])\
     .agg({'AMT_CREDIT_SUM_DEBT': ['count', 'sum', 'mean', 'first'], 'DAYS_CREDIT':['first','last'],\
           'CREDIT_ACTIVE': 'sum' ,'Defaulted':'sum', 'AMT_defaulted':['sum', 'first']})
bt_stat
```

```python
# Flatten last four lags for each credit type
columns   = ['SK_ID_CURR','CREDIT_TYPE','DAYS_CREDIT', 'AMT_CREDIT_SUM_DEBT', 'CREDIT_ACTIVE','Defaulted', 'AMT_defaulted']
bt_stat0  = btype.sort_values('DAYS_CREDIT', ascending = False)[columns].groupby(['SK_ID_CURR','CREDIT_TYPE']).nth(0)
bt_stat1  = btype.sort_values('DAYS_CREDIT', ascending = False)[columns].groupby(['SK_ID_CURR','CREDIT_TYPE']).nth(1)
bt_stat2  = btype.sort_values('DAYS_CREDIT', ascending = False)[columns].groupby(['SK_ID_CURR','CREDIT_TYPE']).nth(2)
bt_stat3  = btype.sort_values('DAYS_CREDIT', ascending = False)[columns].groupby(['SK_ID_CURR','CREDIT_TYPE']).nth(3)
```

```python
ccb.select_dtypes(exclude = 'object').isnull().sum()
```

```python
col = ccb.select_dtypes(exclude='object').columns
ccb[col] = ccb[col].fillna(0)
ccb.isnull().sum()
```

```python
ccb['Beyond_limit'] = np.where(ccb['AMT_BALANCE'] > ccb['AMT_CREDIT_LIMIT_ACTUAL'] , 1, 0)
```

```python
# AMT_INST_MIN_REGULARITY: Minimal installment for this month of the previous credit, 
# AMT_PAYMENT_CURREN: How much did the client pay during the month on the previous credit
ccb['Delinquent'] = np.where(ccb['AMT_PAYMENT_CURRENT'] < ccb['AMT_INST_MIN_REGULARITY'] , 1, 0)
```

```python
# # ccb['balance_pct_change'] = ccb.sort_values('MONTHS_BALANCE').groupby('SK_ID_PREV')['AMT_BALANCE'].pct_change()
def pct_diff(group):
    group['balance_pct_change'] = group['AMT_BALANCE'].diff()/(group['AMT_BALANCE'].shift(1)+1)
    group['balance_pct_change'] = np.where(group['balance_pct_change'] > 30, 30, group['balance_pct_change'])
    return group
ccb = ccb.sort_values('MONTHS_BALANCE').groupby('SK_ID_PREV').apply(pct_diff)           
ccb
```

```python
# fill the NA of first row with zero
ccb['balance_pct_change'].fillna(0, inplace = True)
```

```python
# how many times reached 80% and more
ccb['times_bal80'] = np.where(ccb['AMT_BALANCE']/ccb['AMT_CREDIT_LIMIT_ACTUAL'] >= 0.8, 1, 0)
```

```python
df1 = ccb.groupby(['SK_ID_PREV']).agg({'balance_pct_change': ['mean', 'max', 'last'], 'times_bal80': 'sum'})
df1
```

```python
# How many credit card do applicant have at the moment? What is the total balance on them? (outstanding debt of credit card)
ccb['has_cc_now'] = np.where(ccb['MONTHS_BALANCE'] == -1 , 1 , 0 )
ccb['balance_now'] = ccb['has_cc_now'] * ccb['AMT_BALANCE'] 
df2 = ccb.groupby('SK_ID_CURR').agg({'has_cc_now': 'sum', 'balance_now': 'sum'})
df2
```

# Automated Feature Engineering

```python
# featuretools for automated feature engineering
import featuretools as ft

# ignore warnings from pandas
import warnings
warnings.filterwarnings('ignore')
```

```python
# smaller datasets for automated features
train2     = pd.read_csv('application_train.csv').sort_values('SK_ID_CURR').reset_index().iloc[:1000,:].drop(columns = ['index'])
bureau2    = pd.read_csv('bureau.csv').sort_values(['SK_ID_CURR','SK_ID_BUREAU']).reset_index().iloc[:1000,:].drop(columns = ['index'])
bureaubal2 = pd.read_csv('bureau_balance.csv').sort_values('SK_ID_BUREAU').reset_index().iloc[:1000,:].drop(columns = ['index'])
ccb2       = pd.read_csv('credit_card_balance.csv').sort_values(['SK_ID_CURR','SK_ID_PREV']).reset_index().iloc[:1000,:].drop(columns = ['index'])
insta2     = pd.read_csv('installments_payments.csv').sort_values(['SK_ID_CURR','SK_ID_PREV']).reset_index().iloc[:1000,:].drop(columns = ['index'])
pc2        = pd.read_csv('POS_CASH_balance.csv').sort_values(['SK_ID_CURR','SK_ID_PREV']).reset_index().iloc[:1000,:].drop(columns = ['index'])
prev2      = pd.read_csv('previous_application.csv').sort_values(['SK_ID_CURR','SK_ID_PREV']).reset_index().iloc[:1000,:].drop(columns = ['index'])
```

```python
train2.fillna(0, inplace= True)
bureau2.fillna(0, inplace= True)
bureaubal2.fillna(0, inplace= True)
ccb2.fillna(0, inplace= True)
insta2.fillna(0, inplace= True)
pc2.fillna(0, inplace= True)
prev2.fillna(0, inplace= True)
```

```python
es = ft.EntitySet(id = 'applications')
```

```python
es = es.entity_from_dataframe(entity_id = 'train2'    , dataframe = train2    , index= 'SK_ID_CURR')
es = es.entity_from_dataframe(entity_id = 'bureau2'   , dataframe = bureau2   , index= 'SK_ID_BUREAU')
es = es.entity_from_dataframe(entity_id = 'bureaubal2', dataframe = bureaubal2, make_index = True, index = 'bb_id')
es = es.entity_from_dataframe(entity_id = 'prev2'     , dataframe = prev2     , index = 'SK_ID_PREV')
es = es.entity_from_dataframe(entity_id = 'ccb2'      , dataframe = ccb2      , make_index = True, index = 'cc_id')
es = es.entity_from_dataframe(entity_id = 'insta2'    , dataframe = insta2    , make_index = True, index = 'installment.id')
es = es.entity_from_dataframe(entity_id = 'pc2'       , dataframe = pc2       , make_index = True, index = 'pos_cash_id')
```

```python
# Relationship between application training and bureau
r_applications_bureau = ft.Relationship(es['train2']['SK_ID_CURR'], es['bureau2']['SK_ID_CURR'])
es = es.add_relationship(r_applications_bureau)
```

```python
# Relationship between application training and previous applications
r_app_prev = ft.Relationship(es['train2']['SK_ID_CURR'], es['prev2']['SK_ID_CURR'])
es = es.add_relationship(r_app_prev)
```

```python
# Relationship between previous applications with credit card balance, pos cash, and installments
r_prev_cc = ft.Relationship(es['prev2']['SK_ID_PREV'], es['ccb2']['SK_ID_PREV'])
es = es.add_relationship(r_prev_cc)

r_prev_insta = ft.Relationship(es['prev2']['SK_ID_PREV'], es['insta2']['SK_ID_PREV'])
es = es.add_relationship(r_prev_insta)

r_prev_pc2 = ft.Relationship(es['prev2']['SK_ID_PREV'], es['pc2']['SK_ID_PREV'])
es = es.add_relationship(r_prev_pc2)
```
