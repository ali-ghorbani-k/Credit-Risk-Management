```python
# ratios of different loan status
df = prev.groupby(['SK_ID_CURR','NAME_CONTRACT_STATUS']).agg({'SK_ID_PREV':'count'})
df = df.groupby(level='SK_ID_CURR').apply(lambda x: x/x.sum())
df
```

```python
# flatten the results
approved = df[df.index.isin(['Approved'], level='NAME_CONTRACT_STATUS')]['SK_ID_PREV']
approved.index = approved.index.droplevel('NAME_CONTRACT_STATUS')
refused  = df[df.index.isin(['Refused'], level='NAME_CONTRACT_STATUS')]['SK_ID_PREV']
refused.index = refused.index.droplevel('NAME_CONTRACT_STATUS')
canceled = df[df.index.isin(['Canceled'], level='NAME_CONTRACT_STATUS')]['SK_ID_PREV']
canceled.index = canceled.index.droplevel('NAME_CONTRACT_STATUS')
unused   = df[df.index.isin(['Unused offer'], level='NAME_CONTRACT_STATUS')]['SK_ID_PREV']
unused.index = unused.index.droplevel('NAME_CONTRACT_STATUS')
```

```python
from functools import reduce
data_frames = [approved,refused,canceled,unused]
ratios = reduce(lambda  left,right: pd.merge( left,right, on = 'SK_ID_CURR', how='outer'), data_frames)
ratios.fillna(0, inplace=True)
ratios.columns = ['ratio_approved','ratio_refused','ratio_canceled', 'ratio_unused']
ratios
```

```python
# last application time and contract status
prev1 = prev.sort_values('DAYS_DECISION', ascending = False).\
        groupby(['SK_ID_CURR']).agg({'DAYS_DECISION':'first', 'NAME_CONTRACT_STATUS': 'first', 'AMT_CREDIT' : 'first'})
prev1.head(20)
```

```python
prvs_app = pd.merge(ratios, prev1, on='SK_ID_CURR', how= 'outer')
prvs_app
```

```python
del prev1,ratios, df, approved, refused, canceled, unused, data_frames
```

```python
# latest credit time and amount for each type of contract
prev3 = prev[prev['NAME_CONTRACT_STATUS'] == 'Approved'].sort_values('DAYS_DECISION', ascending = False).\
        groupby(['SK_ID_CURR','NAME_CONTRACT_TYPE']).agg({'DAYS_DECISION':'first', 'AMT_CREDIT' : 'first'})
prev3.head(10)
```

```python
# 'Consumer loans', 'Cash loans', 'Revolving loans'
# flatten the results
consumer_loans = prev3[prev3.index.isin(['Consumer loans'], level='NAME_CONTRACT_TYPE')][['DAYS_DECISION','AMT_CREDIT']]
consumer_loans.index = consumer_loans.index.droplevel('NAME_CONTRACT_TYPE')
cash_loans = prev3[prev3.index.isin(['Cash loans'], level='NAME_CONTRACT_TYPE')][['DAYS_DECISION','AMT_CREDIT']]
cash_loans.index = cash_loans.index.droplevel('NAME_CONTRACT_TYPE')
revolving_loans = prev3[prev3.index.isin(['Revolving loans'], level='NAME_CONTRACT_TYPE')][['DAYS_DECISION','AMT_CREDIT']]
revolving_loans.index = revolving_loans.index.droplevel('NAME_CONTRACT_TYPE')
```

```python
dfs = [consumer_loans,cash_loans, revolving_loans]
cont_type = reduce(lambda left,right: pd.merge(left,right, on ='SK_ID_CURR', how = 'outer'), dfs )
cont_type.columns = ['Days_consumerloans', 'AMT_credit_consumerloans', 'Days_cashloans','AMT_credit_cashloans',
                     'Days_revolvingloans','AMT_credit_revolvingloans']
cont_type.fillna(0, inplace = True)
cont_type
```

```python
prvs_app = pd.merge(prvs_app, cont_type, on='SK_ID_CURR', how= 'outer')
prvs_app
```

```python
del prev3, consumer_loans, cash_loans, revolving_loans, dfs, cont_type
```

```python
ccb = pd.read_csv('credit_card_balance.csv')
ccb
```

```python
# ccb[ccb['SK_ID_CURR'] == 455540]
```

```python
ccb.select_dtypes(exclude = 'object').isnull().sum()
```

```python
col = ccb.select_dtypes(exclude='object').columns
ccb[col] = ccb[col].fillna(0)
ccb.isnull().sum()
```

```python
ccb['Beyond_limit'] = np.where(ccb['AMT_BALANCE'] > ccb['AMT_CREDIT_LIMIT_ACTUAL'] , 1, 0)
```

```python
# AMT_INST_MIN_REGULARITY: Minimal installment for this month of the previous credit, 
# AMT_PAYMENT_CURREN: How much did the client pay during the month on the previous credit
ccb['Delinquent'] = np.where(ccb['AMT_PAYMENT_CURRENT'] < ccb['AMT_INST_MIN_REGULARITY'] , 1, 0)
```

```python
# # ccb['balance_pct_change'] = ccb.sort_values('MONTHS_BALANCE').groupby('SK_ID_PREV')['AMT_BALANCE'].pct_change()
def pct_diff(group):
    group['balance_pct_change'] = group['AMT_BALANCE'].diff()/(group['AMT_BALANCE'].shift(1)+1)
    group['balance_pct_change'] = np.where(group['balance_pct_change'] > 30, 30, group['balance_pct_change'])
    return group
ccb = ccb.sort_values('MONTHS_BALANCE').groupby('SK_ID_PREV').apply(pct_diff)           
ccb
```

```python
# fill the NA of first row with zero
ccb['balance_pct_change'].fillna(0, inplace = True)
```

```python
# how many times reached 80% and more
ccb['times_bal80'] = np.where(ccb['AMT_BALANCE']/ccb['AMT_CREDIT_LIMIT_ACTUAL'] >= 0.8, 1, 0)
```

```python
cols = ['MONTHS_BALANCE','balance_pct_change']
cclast = ccb.sort_values('MONTHS_BALANCE', ascending = False).groupby(['SK_ID_CURR'])[cols].nth(0)
cclag1 = ccb.sort_values('MONTHS_BALANCE', ascending = False).groupby(['SK_ID_CURR'])[cols].nth(1)
cclag2 = ccb.sort_values('MONTHS_BALANCE', ascending = False).groupby(['SK_ID_CURR'])[cols].nth(2)
cclag3 = ccb.sort_values('MONTHS_BALANCE', ascending = False).groupby(['SK_ID_CURR'])[cols].nth(3)
dfs = [cclast, cclag1, cclag2, cclag3]
lags = reduce(lambda left,right: pd.concat([left,right], axis=1, sort= False), dfs)
columns = []
for i in range(4):
    columns += ['MONTHS_BALANCE'+str(i) , 'balance_pct_change'+str(i)]
lags.columns = columns
lags.fillna(0, inplace= True)
lags
```

```python
prvs_app = pd.merge(prvs_app, lags, on='SK_ID_CURR', how= 'outer')
prvs_app
```

```python
del lags, cclast, cclag1, cclag2, cclag3, dfs
```

```python
# How many credit card do applicant have at the moment? What is the total balance on them? (outstanding debt of credit card)
ccb['has_cc_now'] = np.where(ccb['MONTHS_BALANCE'] == -1 , 1 , 0 )
ccb['balance_now'] = ccb['has_cc_now'] * ccb['AMT_BALANCE'] 
df2 = ccb.groupby('SK_ID_CURR', as_index=False).agg({'has_cc_now': 'sum', 'balance_now': 'sum', 'times_bal80': 'sum'})
df2
```

```python
prvs_app = pd.merge(prvs_app, df2, on='SK_ID_CURR', how= 'outer')
```

```python
prvs_app
```

```python
cols = prvs_app.select_dtypes(include = object).columns
prvs_app[cols] = prvs_app[cols].fillna('Missing')
cols = prvs_app.select_dtypes(exclude = object).columns
prvs_app[cols] = prvs_app[cols].fillna(0)
prvs_app.isnull().sum()
```

```python
prvs_app.to_csv('prvs_app.csv', index=False)
```
